```python
import numpy as np

def ReluAct(x):
    return x if x > 0 else 0.0

def derivativeRelu(x):
    return 1.0 if x > 0 else 0.0

inputs = np.array([10.0, 20.0, 30.0, 40.0])   
weights = np.random.randn(3, 4)           
biases = np.zeros(3)
target = 5.0
lr = 0.01

w11, w12, w13, w14 = weights[0]
w21, w22, w23, w24 = weights[1]
w31, w32, w33, w34 = weights[2]

dy_da1 = 1.0
dy_da2 = 1.0
dy_da3 = 1.0

for it in range(100):
    z1 = inputs[0]*w11 + inputs[1]*w12 + inputs[2]*w13 + inputs[3]*w14 + biases[0]
    z2 = inputs[0]*w21 + inputs[1]*w22 + inputs[2]*w23 + inputs[3]*w24 + biases[1]
    z3 = inputs[0]*w31 + inputs[1]*w32 + inputs[2]*w33 + inputs[3]*w34 + biases[2]

    a1 = ReluAct(z1)
    a2 = ReluAct(z2)
    a3 = ReluAct(z3)

    y = a1 + a2 + a3
    loss = (y - target)**2

    dL_dy = 2 * (y - target)

    da1_dz1 = derivativeRelu(z1)
    da2_dz2 = derivativeRelu(z2)
    da3_dz3 = derivativeRelu(z3)

    dz1_dw11 = inputs[0]; dz1_dw12 = inputs[1]; dz1_dw13 = inputs[2]; dz1_dw14 = inputs[3]
    dz2_dw21 = inputs[0]; dz2_dw22 = inputs[1]; dz2_dw23 = inputs[2]; dz2_dw24 = inputs[3]
    dz3_dw31 = inputs[0]; dz3_dw32 = inputs[1]; dz3_dw33 = inputs[2]; dz3_dw34 = inputs[3]

    dL_dw11 = dL_dy * dy_da1 * da1_dz1 * dz1_dw11
    dL_dw12 = dL_dy * dy_da1 * da1_dz1 * dz1_dw12
    dL_dw13 = dL_dy * dy_da1 * da1_dz1 * dz1_dw13
    dL_dw14 = dL_dy * dy_da1 * da1_dz1 * dz1_dw14

    dL_dw21 = dL_dy * dy_da2 * da2_dz2 * dz2_dw21
    dL_dw22 = dL_dy * dy_da2 * da2_dz2 * dz2_dw22
    dL_dw23 = dL_dy * dy_da2 * da2_dz2 * dz2_dw23
    dL_dw24 = dL_dy * dy_da2 * da2_dz2 * dz2_dw24

    dL_dw31 = dL_dy * dy_da3 * da3_dz3 * dz3_dw31
    dL_dw32 = dL_dy * dy_da3 * da3_dz3 * dz3_dw32
    dL_dw33 = dL_dy * dy_da3 * da3_dz3 * dz3_dw33
    dL_dw34 = dL_dy * dy_da3 * da3_dz3 * dz3_dw34

    w11 -= lr * dL_dw11
    w12 -= lr * dL_dw12
    w13 -= lr * dL_dw13
    w14 -= lr * dL_dw14

    w21 -= lr * dL_dw21
    w22 -= lr * dL_dw22
    w23 -= lr * dL_dw23
    w24 -= lr * dL_dw24

    w31 -= lr * dL_dw31
    w32 -= lr * dL_dw32
    w33 -= lr * dL_dw33
    w34 -= lr * dL_dw34

    db1 = dL_dy * dy_da1 * da1_dz1 * 1
    db2 = dL_dy * dy_da2 * da2_dz2 * 1
    db3 = dL_dy * dy_da3 * da3_dz3 * 1
    biases[0] -= lr * db1
    biases[1] -= lr * db2
    biases[2] -= lr * db3

    
    print(f"iter {it:03d}  loss {loss:.6f}  y {y:.4f}")
```

```python
import numpy as np

def ReluAct(x):
    return np.maximum(0, x)

def derivativeRelu(x):
    return (x > 0).astype(float)

inputs = np.array([10.0, 20.0, 30.0, 40.0])   
weights = np.random.randn(3, 4)              
biases = np.zeros(3)                        
target = 5.0
lr = 0.001


dy_da = np.ones(3)   

for it in range(100):

    z = weights.dot(inputs) + biases        
    a = ReluAct(z)                        
    y = a.sum()                           
    loss = (y - target)**2

    dL_dy = 2.0 * (y )             

    da_dz = derivativeRelu(z)              
    dL_da = dL_dy * dy_da                  
    dL_dz = dL_da * da_dz                 

    dL_dW = dL_dz[:, np.newaxis] * inputs[np.newaxis, :]   

    dL_db = dL_dz.copy()                  

    weights -= lr * dL_dW
    biases  -= lr * dL_db

    if it % 10 == 0:
        print(f"iter {it:03d}  loss {loss:.6f}  y {y:.4f}")
```
